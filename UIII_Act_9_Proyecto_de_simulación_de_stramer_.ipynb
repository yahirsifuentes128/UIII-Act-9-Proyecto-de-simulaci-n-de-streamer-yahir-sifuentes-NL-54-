```python

# IA para Clase Streamer

print("===================================")
print(" IA - Clase Streamer")
print("===================================")

while True:

    print("\n========= MENÚ =========")
    print("1. Analizar streamer")
    print("2. Ver consejos")
    print("3. Salir")

    opcion = input("Seleccione una opción: ")

    # Opción 1
    if opcion == "1":

        nombre = input("\nNombre del streamer: ")
        seguidores = int(input("Cantidad de seguidores: "))
        horas = int(input("Horas de transmisión: "))
        contenido = input("Contenido atractivo (si/no): ")
        reportes = int(input("Cantidad de reportes: "))

        print("\nAnalizando canal...\n")

        # Simulación de IA
        if seguidores >= 50000 and horas >= 20 and contenido == "si" and reportes == 0:
            print(nombre, "tiene un canal EXITOSO")
            print("Nivel del canal: ALTO")

        elif seguidores >= 10000 and contenido == "si" and reportes <= 2:
            print(nombre, "tiene un canal EN CRECIMIENTO")
            print("Nivel del canal: MEDIO")

        else:
            print(nombre, "tiene un canal CON PROBLEMAS")
            print("Nivel del canal: BAJO")

    # Opción 2
    elif opcion == "2":

        print("\n===== CONSEJOS PARA STREAMERS =====")
        print("- Hablar con los espectadores")
        print("- Tener horarios fijos")
        print("- Crear contenido divertido")
        print("- Evitar lenguaje ofensivo")
        print("- Usar buena iluminación")

    # Opción 3
    elif opcion == "3":

        print("\nPrograma realizado por Yahir Sifuentes NL = 54.")
        break

    # Opción inválida
    else:
        print("\nOpción no válida.")
```
