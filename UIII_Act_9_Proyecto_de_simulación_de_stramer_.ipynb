```python

# IA para Clase Streamer
# Lista vacía al inicio

streamers = []

print("===================================")
print(" IA - Clase Streamer")
print(" CBTIS 128 - Proyecto IA")
print("===================================")

while True:

    print("\n========= MENÚ =========")
    print("1. Registrar streamer")
    print("2. Analizar streamer")
    print("3. Ver lista de streamers")
    print("4. Ver consejos")
    print("5. Salir")

    opcion = input("Seleccione una opción: ")

    # OPCIÓN 1
    if opcion == "1":

        nombre = input("\nNombre del streamer: ")

        # Agregar a la lista
        streamers.append(nombre)

        print(nombre, "fue agregado correctamente.")
        print("Lista actual:", streamers)

    # OPCIÓN 2
    elif opcion == "2":

        if streamers == []:
            print("\nNo hay streamers registrados.")

        else:

            nombre = input("\nNombre del streamer a analizar: ")

            if nombre in streamers:

                seguidores = int(input("Cantidad de seguidores: "))
                horas = int(input("Horas de transmisión: "))
                contenido = input("Contenido atractivo (si/no): ")
                reportes = int(input("Cantidad de reportes: "))

                print("\nAnalizando canal...\n")

                # Simulación de IA
                if seguidores >= 50000 and horas >= 20 and contenido == "si" and reportes == 0:

                    print(nombre, "tiene un canal EXITOSO")
                    print("Nivel del canal: ALTO")

                elif seguidores >= 10000 and contenido == "si" and reportes <= 2:

                    print(nombre, "tiene un canal EN CRECIMIENTO")
                    print("Nivel del canal: MEDIO")

                else:

                    print(nombre, "tiene un canal CON PROBLEMAS")
                    print("Nivel del canal: BAJO")

            else:
                print("\nEse streamer no está registrado.")

    # OPCIÓN 3
    elif opcion == "3":

        print("\n===== LISTA DE STREAMERS =====")

        if streamers == []:
            print("La lista está vacía.")

        else:
            print(streamers)

    # OPCIÓN 4
    elif opcion == "4":

        print("\n===== CONSEJOS PARA STREAMERS =====")
        print("- Hablar con los espectadores")
        print("- Tener horarios fijos")
        print("- Crear contenido divertido")
        print("- Evitar lenguaje ofensivo")
        print("- Usar buena iluminación")

    # OPCIÓN 5
    elif opcion == "5":

        print("\nPrograma realizado por Yahir Sifuentes Vargas")
        print("NL = 54")
        break

    # OPCIÓN INVÁLIDA
    else:

        print("\nOpción no válida.")
```
